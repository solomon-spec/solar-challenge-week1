```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in numerical_cols:
    if col == 'Comments':
        continue
    print(f"\nAnalyzing: {col}")
    print(df[col].describe()) # Print descriptive statistics for the current numerical column

    plt.figure(figsize=(18, 5)) # Create a figure for three subplots

    # Subplot 1: Histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[col], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)

    # Subplot 2: Box Plot (for outlier detection and spread)
    plt.subplot(1, 3, 2)
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f'Box Plot of {col}', fontsize=14)
    plt.ylabel(col, fontsize=12)

    # Subplot 3: Density Plot (KDE - smoothed distribution)
    plt.subplot(1, 3, 3)
    sns.kdeplot(df[col], fill=True, color='lightgreen')
    plt.title(f'Density Plot of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Density', fontsize=12)

    plt.tight_layout() # Adjust subplot parameters for a tight layout
    plt.show() # Display the plots for the current numerical column
```

```python
print("\n" + "="*30)
print("--- Bivariate Analysis ---")
print("="*30)

print("\n--- Numerical vs. Numerical Relationships (Scatter Plots) ---")

# Select a few key numerical pairs for scatter plots
# GHI vs DNI: Expect a strong positive correlation
plt.figure(figsize=(10, 7))
sns.scatterplot(x='GHI', y='DNI', data=df, alpha=0.7, color='purple')
plt.title('GHI vs. DNI', fontsize=16)
plt.xlabel('GHI (W/m²)', fontsize=12) # Add units back to labels for clarity
plt.ylabel('DNI (W/m²)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Tamb vs TModA: Expect a strong positive correlation
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Tamb', y='TModA', data=df, alpha=0.7, color='teal')
plt.title('Ambient Temperature vs. Module A Temperature', fontsize=16)
plt.xlabel('Ambient Temperature (°C)', fontsize=12)
plt.ylabel('Module A Temperature (°C)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# WS vs WSgust: Expect a strong positive correlation
plt.figure(figsize=(10, 7))
sns.scatterplot(x='WS', y='WSgust', data=df, alpha=0.7, color='darkorange')
plt.title('Wind Speed vs. Wind Gust Speed', fontsize=16)
plt.xlabel('Wind Speed (m/s)', fontsize=12)
plt.ylabel('Wind Gust Speed (m/s)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
print("\n--- Correlation Matrix of Numerical Features ---")
# Create a DataFrame containing only numerical columns for correlation calculation
numerical_for_corr = df[numerical_cols]
correlation_matrix = numerical_for_corr.corr()

plt.figure(figsize=(16, 14)) # Adjust figure size for better readability of labels
# Create a heatmap to visualize the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, annot_kws={"size": 10})
plt.title('Correlation Matrix of Numerical Features', fontsize=18)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
