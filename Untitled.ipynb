```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def plot_classification_metrics(metrics_data):
    """
    Plots various classification metrics as a bar chart.

    Args:
        metrics_data (dict): A dictionary where keys are metric names
                             and values are their corresponding numerical values.
                             Example: {'Accuracy': 0.7278, 'Precision': 0.7306, ...}
    """

    # Convert the dictionary to a pandas DataFrame for easier plotting
    df = pd.DataFrame(list(metrics_data.items()), columns=['Metric', 'Value'])

    # Sort the DataFrame by 'Value' for better visualization, ascending for loss/brier, descending for others
    # For simplicity, we'll sort all by value in descending order, but one could customize this.
    # It's usually good to group similar metrics (e.g., all accuracies, all F1s)
    
    # Define categories for better grouping and sorting
    performance_metrics = ['Accuracy', 'Balanced Accuracy', 'Macro Avg Precision', 'Macro Avg Recall',
                           'Macro Avg F1-Score', 'Weighted Avg Precision', 'Weighted Avg Recall',
                           'Weighted Avg F1-Score', 'Micro Avg Precision', 'Micro Avg Recall',
                           'Micro Avg F1-Score', 'Matthews Correlation Coefficient (MCC)', "Cohen’s Kappa"]
    
    error_metrics = ['Log Loss (Cross-Entropy)', 'Brier Score (Multiclass)']

    # Separate and sort
    df_performance = df[df['Metric'].isin(performance_metrics)].sort_values(by='Value', ascending=False)
    df_error = df[df['Metric'].isin(error_metrics)].sort_values(by='Value', ascending=True)

    # Combine them, putting performance first then error metrics
    df_sorted = pd.concat([df_performance, df_error])

    plt.figure(figsize=(14, 8)) # Adjust figure size for better readability
    sns.barplot(x='Value', y='Metric', data=df_sorted, palette='viridis')

    plt.title('Classification Model Performance Metrics', fontsize=16)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Metric', fontsize=12)
    plt.xlim(0, 1.1) # Set x-axis limit from 0 to slightly above max value for better scale

    # Add value labels on the bars
    for index, row in df_sorted.iterrows():
        plt.text(row['Value'] + 0.01, index, f"{row['Value']:.4f}", color='black', ha="left", va='center', fontsize=10)

    plt.grid(axis='x', linestyle='--', alpha=0.7) # Add grid lines for better readability
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()

# Data provided by the user
metrics_data = {
    'Accuracy': 0.7278,
    'Balanced Accuracy': 0.7269,
    'Macro Avg Precision': 0.7306,
    'Macro Avg Recall': 0.7269,
    'Macro Avg F1-Score': 0.7259,
    'Weighted Avg Precision': 0.7331,
    'Weighted Avg Recall': 0.7278,
    'Weighted Avg F1-Score': 0.7276,
    'Micro Avg Precision': 0.7278,
    'Micro Avg Recall': 0.7278,
    'Micro Avg F1-Score': 0.7278,
    'Matthews Correlation Coefficient (MCC)': 0.7251,
    'Log Loss (Cross-Entropy)': 1.0366,
    "Cohen’s Kappa": 0.7251,
    'Brier Score (Multiclass)': 0.3713
}

# Call the function to plot the metrics
plot_classification_metrics(metrics_data)
```
